```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mglearn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from numpy import genfromtxt
from pandas.tools.plotting import scatter_matrix
from sklearn.neighbors import KNeighborsClassifier
import csv

%matplotlib inline
def main():
    
    iris_dataset =  pd.read_csv('dataset/iris.csv', header=0)
    
    #convert species to numeric value 
    mapping = {'Iris-virginica': 1, 'Iris-setosa': 2, 'Iris-versicolor' : 3 }
    iris_dataset = iris_dataset.replace({'Species' : mapping})
    
    #target
    target = iris_dataset['Species']
            
    #data
    data = iris_dataset[ ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'] ]
    
    X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=0)
    
    grr = pd.scatter_matrix(X_train, c=y_train, figsize=(15,15), 
                            marker='o', hist_kwds={'bins' : 20}, s=60, alpha=.8, cmap=mglearn.cm3)
    
        
    knn = KNeighborsClassifier(n_neighbors=1)
    knn.fit(X_train, y_train)
    
    y_predict = knn.predict(X_test)
    
    print('test score ::: \n{}'.format(knn.score(X_test,y_test)))

if __name__ == '__main__':
    main()
```
